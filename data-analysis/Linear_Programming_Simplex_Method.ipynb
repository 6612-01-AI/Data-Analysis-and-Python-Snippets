```python
from termcolor import colored     #to style the text
import numpy as np
from prettytable import PrettyTable     #to form the table in simplex algorithm
from fractions import Fraction # to define the values as fractions instead of decimal

#definition of the function
def que(A_coef,b,c,B,cb,n,v):

    st_a=[]   #array where row operations for each table is stored
    xb = np.transpose([b])

    # combining matrix B,cb,xb,A in the table
    table = np.hstack((B, cb))
    table = np.hstack((table, xb))
    table = np.hstack((table, A_coef))

    #this forms the whole table
    table = np.array(table, dtype ='float')

    MIN = 0   #since, this is a min problem

    print (colored('\n Simplex Algorithm:\n', attrs=['bold']))


    obtained = 0  #to check the optimality condition
    iteration = 1      #to keep the count on no. of iterations

    while obtained == 0:
        #print(st_a)

        simplex_t= PrettyTable()    #initializing the table for simple algorithm

        comb=["Xb","B","Cb","b"]     #defining the header columns

        for i in range(n+v):
            j=i+1
            temp = 'a' + '%d' %j
            comb.append(temp)
        comb.append("Min ratio")
        comb.append("Operations")
        simplex_t=PrettyTable(comb)       #combining the header columns

        print("The iteration no is:",iteration,"\n")

        i = 0
        z_c = []
        temp3=[]

        while i<len(A_coef[0]):
            z_c.append(c[i] - np.sum(table[:, 1]*table[:, 3 + i])) #to find ci-zj
            i = i + 1

        i = 0
        b_var = table[:, 0]    #basic variable array

        check = 0
        for zicj in z_c:
            if zicj>0:
                check = 1     #to find if all the values of zi-cj are positive

         #adding to the last the value of Cj for each variable which remains constant
        temp5=[]
        #temp5.append(" ")
        temp5.append("-")
        temp5.append("-")
        temp5.append("-")
        temp5.append(colored('Cj', attrs=['bold']))
        for g in range(n+v):
            temp5.append(colored(c[g], attrs=['bold']))
        temp5.append(" ")
        temp5.append(" ")
        simplex_t.add_row(temp5)  #adding the extra row containing Cjs


        k = z_c.index(max(z_c)) #variable entering the basis
        min = 99999
        f = 0;
        r = -1
        temp5=[]
        var=0
        num=0

        for row in table:
            ing=0
            temp2=[]
            for el in row:
                var2=Fraction(str(el)).limit_denominator(100)     #defining each row in the table by variable coefficients
                if(ing==0):     #a flag
                    #temp2.append(var2+1)      #defining the indices of basic variables
                    s1=("X%d"%(var2+1))
                    s2=("a%d"%(var2+1))
                    temp2.append(s1)
                    temp2.append(s2)
                else:
                    temp2.append(var2)
                ing=ing+1

            var=var+temp2[2]*temp2[3] #calculating the optimal value in each operation

            #min ratio calculation
            if (table[:, 2][f]>0 and table[:, 3 + k][f]>0):
                val = table[:, 2][f]/table[:, 3 + k][f]
                value=Fraction(val).limit_denominator(100)
                if(check==0):
                    temp2.append("--")
                else:
                    temp2.append(value)
                if val<min:
                    min = val
                    r = f # leaving variable

            else:
                temp2.append("--")
            f=f+1
            if(iteration!=1):
                temp2.append(st_a[num])
                num=num+1;
            else:
                temp2.append(" ")
            simplex_t.add_row(temp2) #addition of a row


        #appending the final row which contain value of Zi-Cj
        #temp3.append(" ")
        temp3.append(" ")
        temp3.append(" ")
        temp3.append("Zj-Cj")
        temp3.append(var)


        for zicj in z_c:
            var3=-Fraction(str(zicj)).limit_denominator(100)
            temp3.append(var3)

        temp3.append(" ")
        temp3.append(" ")

        simplex_t.add_row(temp3)     #adding the final row



        print(simplex_t,"\n")   #printing the simplex table



        check = 0
        for zicj in z_c:
            if zicj>0:
                check = 1
                break

        if check == 0:
            print("All Zi-Cj are >= 0. So, the solution is optimally reached")
            obtained = 1
            break


        pivot = table[r][3 + k]    #finding the pivot element
        piv=Fraction(pivot).limit_denominator(100)   #to convert the pivot element to fraction

        table[r, 2:len(table[0])] = table[ r, 2:len(table[0])] / pivot
        st_a=[]    #stores the row operations for each table

        i = 0

        #the required row-operations for the pivot element
        while i<len(table):
            if i != r:
                mul=Fraction(str(table[i][3+k])).limit_denominator(100)
                #to make the value 0 of pivot variable
                table[i, 1:len(table[0])] = table[i,
                     1:len(table[0])] - table[i][3 + k] * table[r, 1:len(table[0])]

                st=("R'%d=R%d-(%s)*R'%d"%(i+1,i+1,mul,r+1))    #storing the row operations as strings
                st_a.append(st)
            else:
                st1=("R'%d=R/(%s)"%(r+1,piv))    #storing the row operation for pivot to make it 1
                st_a.append(st1)
                #print(st)
            i += 1


        # assign the entering new basic variable
        table[r][0] = k
        table[r][1] = c[k]
        iteration+= 1


    basis = []
    i = 0
    sum = 0
    while i<len(table):
        sum += c[int(table[i][0])]*table[i][2]
        temp = "x"+str(int(table[i][0])+1)
        basis.append(temp)   #to find the basis that contains the variable in final solution
        i+= 1

    ans=Fraction(str(sum)).limit_denominator(100)   #to find the optimal value



    be=[]

    for k in range(n):
        be.append(Fraction(str(table[k][2])).limit_denominator(100))   #to calculate the value of basis variable
    temp4=[]


    print (colored('\nOptimal solution is: \n', attrs=['bold']))

    #to check the main variables are in optimal solution
    for i in range(v):
        k=i+1
        flg=0
        var3 = 'x' + '%d' %k
        for j in range(len(basis)):
            if(var3==basis[j]):    #when the main variable appears in the basis
                flg=1
                print(var3, "=",be[j],end=', ')
                break;
        if(flg==0):
            print(var3, "=",0,end=', ')   #when the main variable is not in the basis

    print("and Z =",colored(ans, attrs=['bold']))     #the optimal value


```
